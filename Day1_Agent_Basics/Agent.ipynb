```python
from google.adk import Agent, tool
from google import genai
from dotenv import load_dotenv
import os
```

```python
# Load API key
load_dotenv()
genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))
```

```python
# Define the Google Search tool
@tool
def google_search(query: str):
    """Searches Google and returns relevant results."""
    # This is just a mock example ‚Äî you can later connect real APIs here.
    return f"üîç Pretending to search Google for: {query}"
```

```python
# Define the root agent
root_agent = Agent(
    name="helpful_assistant",
    model="gemini-2.5-flash",
    description="A simple agent that can answer general questions.",
    instruction="You are a helpful assistant. Use Google Search for current info or if unsure.",
    tools=[google_search],
)

print("‚úÖ Root Agent defined.")
```

```python
# Example query
response = root_agent.run("Who won the FIFA World Cup in 2022?")
print("ü§ñ Agent:", response.text)
```
